```python
import numpy as np
import matplotlib.pyplot as plt
import math # Added import for math module

lambda1=4
N=100000
av=np.array([])

for j in range(N):
    u=np.random.uniform(0,1)
    p=np.exp(-lambda1)
    i=0
    F=p

    while u>=F:
        p=(lambda1/(i+1))*p
        F=F+p
        i=i+1

    av=np.append(av,i)

print("Media simulada =",np.mean(av))
print("Variancia simulada =",np.var(av))

# probabilidades estimadas por simulacao
p6=np.sum(av==6)/N
p2=np.sum(av<=2)/N

print("P(X=6) simulada =",p6)
print("P(X<=2) simulada =",p2)

# em 2 horas -> lambda = 8
lambda2=8
av2=np.array([])

for j in range(N):
    u=np.random.uniform(0,1)
    p=np.exp(-lambda2)
    i=0
    F=p

    while u>=F:
        p=(lambda2/(i+1))*p
        F=F+p
        i=i+1

    av2=np.append(av2,i)

pmaior5=np.sum(av2>5)/N
print("P(Y>5) simulada com lambda=8 =",pmaior5)

X=np.arange(0,15,1)
fx=np.array([])

for k in X:
    fx=np.append(fx,(lambda1**k)*np.exp(-lambda1)/math.factorial(k)) # Corrected to use math.factorial

plt.plot(X,fx)
plt.hist(av,bins=np.arange(-0.5,15.5,1),density=True)
plt.show()
```
