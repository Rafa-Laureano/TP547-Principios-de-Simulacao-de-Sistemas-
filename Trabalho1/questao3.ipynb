```python
import numpy as np
import matplotlib.pyplot as plt
import math # Added import for math module

q=0.08
n=20
N=100000
c=q/(1-q)

av=np.array([])

for j in range(N):
    u=np.random.uniform(0,1)
    i=0
    pr=pow((1-q),n)
    F=pr

    while u>=F:
        pr=(c*(n-i)/(i+1))*pr
        F=F+pr
        i=i+1

    av=np.append(av,i)

print("Media simulada =",np.mean(av))
print("Variancia simulada =",np.var(av))

print("P(X=3) simulada =",np.sum(av==3)/N)
print("P(X<=2) simulada =",np.sum(av<=2)/N)
print("P(X>5) simulada =",np.sum(av>5)/N)

X=np.arange(0,n+1,1)
fx=np.array([])

for k in X:
    fx=np.append(fx,math.factorial(n)/(math.factorial(k)*math.factorial(n-k))*pow(q,k)*pow((1-q),(n-k))) # Corrected to use math.factorial

plt.plot(X,fx)
plt.hist(av,bins=np.arange(-0.5,n+1.5,1),density=True)
plt.show()
```
