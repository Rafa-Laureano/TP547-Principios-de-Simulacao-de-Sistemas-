```python
import numpy as np
import matplotlib.pyplot as plt

a=11
c=7
m=32
x=5

x1=np.array([x])

n=16
for i in range(n):
    x=(a*x+c)%m
    x1=np.append(x1,x)

print("Sequencia gerada:")
print(x1)

print("10 primeiros numeros apos X0:")
print(x1[1:11])

# periodo
x=5
count=0
while True:
    x=(a*x+c)%m
    count=count+1
    if x==5:
        break

print("Periodo =",count)

ind=np.arange(np.size(x1))
plt.bar(ind,x1)
plt.xlabel('amostra')
plt.ylabel('valor gerado')
plt.title('Gerador Linear Congruente')
plt.show()
```
