```python
import numpy as np
import matplotlib.pyplot as plt

p=0.3
n=100000
xfunction=np.array([])

for i in range(n):
    count=0
    u=np.random.uniform(0,1)

    while u>p:
        count=count+1
        u=np.random.uniform(0,1)

    xfunction=np.append(xfunction,count)

print("Media simulada =",np.mean(xfunction))
print("Variancia simulada =",np.var(xfunction))

print("P(primeiro sucesso na 4a tentativa) simulada =",np.sum(xfunction==3)/n)
print("P(mais de 6 tentativas) simulada =",np.sum(xfunction>=6)/n)

X=np.arange(0,15,1)
fx=(1-p)**X*p

plt.plot(X,fx)
plt.hist(xfunction,bins=np.arange(-0.5,15.5,1),density=True)
plt.show()
```
